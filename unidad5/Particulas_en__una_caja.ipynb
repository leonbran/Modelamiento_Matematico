# Solución del problema de las partículas en una caja

```python
#Importar módulo numpy para introducir la librería random
import numpy as np 

N_particulas=100 #numero total de particulas
t_max = 10*N_particulas #Tiempo de convergencia

#Ponga todas las particulas en la caja de la izquierda
#n_izq: particulas de la izquierda
#N-n_izq: particulas de la derecha
n_izq = N_particulas

#Abrir un archivo para escritura
file=open('particulas.txt','w')

#ciclo temporal
for t in range(0,t_max):
    aleatorio = np.random.randint(1,N_particulas)

    if (aleatorio <= n_izq):
        n_izq -= 1
    else:
        n_izq += 1

#Escribir la informacion en el archivo 
    file.write("%i %i\n" %(n_izq,N_particulas-n_izq))

#Cerrar el archivo
file.close()

#importar módulo matplotlib.pyplot para realizar gráficos
#se grafican las particulas de la izquierda y derecha en funcion del tiempo
#con la informacion que esta en el archivo
import matplotlib.pyplot as plt

x= np.loadtxt('particulas.txt')
plt.ylim([0,100])
plt.xlabel("Tiempo")
plt.plot(x)
plt.show()
```
