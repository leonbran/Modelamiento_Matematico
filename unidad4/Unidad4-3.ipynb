# Calcula recta de regresion lineal que pasa por el origen 

```python
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
import numpy as np

x = np.array( [ 9.6 , 18.3 , 29, 47.2 , 71.1 , 119.1 , 174.6 ] )
y = np.array( [8.7 , 10.7 , 18.2 , 23.9 , 48.0 , 55.5 , 82.7 ] )

def linealreg(x, k):
  return k*x


# Calcula los parametros de la regresion lineal
k = curve_fit( linealreg , x, y)
print(k[0]) # Imprime el valor de la pendiente de la recta de regresion

# Grafica resultados
plt.plot(x, linealreg(x, k[0] ),'k' )
plt.plot(x,y,'ko ')
plt.xlabel(r'$P_n$', size = 14)
plt.text(2, 80, r'$\Delta P_n$', size = 14)
plt.show()

```
