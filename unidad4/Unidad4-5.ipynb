# Gráfica campo de dirección de una ecuación diferencial

```python
import numpy as np
import matplotlib.pyplot as plt


# Define rango del eje temporal
a = 0
b = 7


t = np.arange(a, b + 0.5 , .5)
y = np.arange(-1, 1.5 , .2)
T, Y = np.meshgrid(t, y)


def f(t, y):
  return 3*y

dYdT = f(T, Y)
U = (1/(1 + dYdT**2)**.4)*np.ones(T. shape )
V = (1/(1 + dYdT**2)**.4)*dYdT

# Ajusta dimensiones del grafico
fig = plt.figure( figsize = (10 ,8) )

# Realiza el grafico
plt.quiver(T, Y, U, V) # Grafica campo de direcciones
plt.plot( [0, 7], [0, 0] ,'k--') # Grafica solucion de equilibrio y = 0
plt.show()
```
