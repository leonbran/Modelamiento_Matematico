# Aproxima la solución de un problema de valor inicial asociado a una ecuación diferencial

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')


# Rutina para evaluar la funcion f(y,t) que define la EDO
def model(y,t):
  dydt = 3*y
  return dydt


t = np.linspace(0 , 0.5) # Definicion del vector de tiempos
y0 = 1 # Condicion inicial
y = odeint(model , y0 , t ) # Solucion numerica de la EDO


# Graficando resultados
fig = plt.figure()
ax = plt.axes()
ax.plot(t, y, 'b')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.ylim(-10 ,10)
plt.show()
```
