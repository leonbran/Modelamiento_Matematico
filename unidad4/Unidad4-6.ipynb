# Implementa el método de Euler para aproximar una solución de una ecuación diferencial

```python

import numpy as np
import matplotlib.pyplot as plt

def F(t, y):
  return 1 + (y-t)**2

def exacta (t):
  return t + 1/(2 - t)

T= 1.9
N = 50
t = np.linspace(0, T, N+1)
h = t[2] - t[1]
print("Tamaño de paso :", h)   # Imprime el tamano de paso usado
y = np.empty(N+1)

y[0] = 0.5 # Define la condicion inicial

for i in range(0, N ):
  y[i+1] = y[i] + h*F(t[i],y[i])

fig = plt.figure( figsize =(15 ,8) )
plt.plot(t, y,'--k')
plt.plot(t, exacta(t),'.k' )
plt.xlabel(r'$t$', size = 16)
plt.text(0, 12, r'$y(t)$', size = 16)
plt.grid()
plt.show()
```
