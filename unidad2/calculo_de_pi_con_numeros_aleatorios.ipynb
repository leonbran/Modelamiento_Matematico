# Cálculo del número $\pi$ con números aleatorios

**Descripción:** En este programa calculamos el valor del número $\pi$ utilizando números aleatorios de la siguiente manera:

Considere un círculo de diámetro $L$ inscrito dentro de un cuadrado de lado $L$ como se muestra en la siguiente figura:

Dado que el cociente de sus áreas es 


\begin{equation*}
\dfrac{\text{Área Circulo}}{\text{Área Cuadrado}} = \dfrac{\pi \left(\dfrac{L}{2}\right)^2}{L^2} = \dfrac{\pi}{4},
\end{equation*}



el valor de $\pi$ puede aproximarse de la siguiente forma:

* Escoga un número $n$ de puntos aleatorios dentro del cuadrado.

* Cuente el número de puntos dentro del círculo.

* El cociente del número de puntos dentro del círculo dividido entre el número $n$ es aproximadamente $\pi/4$. 


**Nota:** Entre mas grande sea $n$, mejor sera nuestra aproximación del numero $\pi$.


```python
#importando el módulo numpy
import numpy as np


#-------------Se define la función de cálculo de Pi---------------

def Calculo_de_Pi(n):

        
    PuntosDentroCirculo = 0

    ContadorCiclo = 0

    #Se hace un ciclo hasta que contador sea igual al limite

    while  ContadorCiclo < n:

        #Se calcula  que los puntos esten dentro del radio del circulo

        #Si es asi se incrementa la variable dentro. y luego se incrementa

        #contador para pasar al siguiente ciclo.

        x = np.random.rand()
        
        y = np.random.rand() 
        
        if ( ( x**2 + y**2) <= 1 ):

            PuntosDentroCirculo += 1

        ContadorCiclo += 1

    #Se retorna el resultado de la cantida de puntos dentro del círculo
    return 4.0 * (PuntosDentroCirculo / n)
#------------------------------------------------------------------------


#Escoger varios números n de intentos para hacer los experimentos
lista_de_n = [100, 10000 , 1000000]

for i in lista_de_n :

    print ("El valor de pi es %f para n igual a %i \n " %(Calculo_de_Pi(i),i))

```

#### Cálculo del error numérico con respecto al número de intentos (opcional)

```python
#Graficando el error con respecto al numero de pasos desde 100 hasta 10.000 intentos

#Tomamos como valor exacto el correspondiente al modulo numpy
ValorExactoPi = np.pi

#Genero dos listas vacias para almacenar los errores y los intentos
Listas_de_Errores = []
Lista_de_n        = []

for i in range( 100, 10000 , 100):
    
        Intento = i
        Error   = abs( np.pi - Calculo_de_Pi(i) )
        
        #Adicionando "Intento" a lista de intentos
        Lista_de_n.append( Intento )
        
        #Adicionando "Error" a lista de errores
        Listas_de_Errores.append( Error    )
       
        
import matplotlib.pyplot as plt


#graficando las listas. 
plt.plot(Lista_de_n,Listas_de_Errores)


#Colocando escala logaritmica en el eje y
plt.yscale("log")


#poniendo nombres a de los ejes
plt.xlabel('n')
plt.ylabel('Error')


# ajustar tamaño de la figura (ancho y alto)
plt.rcParams['figure.figsize'] = [7, 5]
 
#Mostrando el grafico
plt.show()
```
