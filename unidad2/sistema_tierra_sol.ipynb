# Gráfico de la trayectoria de la tierra 

Graficamos la  trayectoria de la tierra al rededor del sol, la cual esta dada por la siguiente función en coordenadas polares $$r(\theta) = \frac{d}{1+ \epsilon cos \theta}$$ para diferentes valores del parametro de excentricidad $\epsilon$.  

```python
import numpy as np
import matplotlib.pyplot as plt


def radio(theta , e , d ):        
    return d /( 1+ e*np.cos(theta) )

lista_thetas = np.linspace(0, 2*np.pi , 100)

d     = 1

#vamos a explorar varios graficos
lista_e = [  0.8 , 1.0 , 1.5]

for e in lista_e:

    x = []
    y = []

    for theta in lista_thetas:

        denominador = 1+e*np.cos(theta) 

        if denominador != 0:

            r = d / denominador

            x.append( r * np.cos(theta) )
            y.append( r * np.sin(theta) )

        else:
            continue

    plt.plot(x,y,label= "e = " +str(e))
    plt.legend()
    plt.grid()
    plt.xlim(-10,5)
    plt.ylim(-5,5)

plt.show()
```
