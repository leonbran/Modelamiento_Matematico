# Calcula regresión exponencial

```python
# Importa librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define la función para mínimos cuadrados
def func(x,a,b,c):
  return a*x**3 + b*np.exp(c*x)

# Define los datos
xdata = np.array([-2.18,-1.527,-0.61,0.84,1.75,2.38])
ydata = np.array([14.765,-3.142,-9.252,-10.342,-2.371,19.705])

# Calcula la curva de regresión
popt, _ = curve_fit(func, xdata, ydata)

# Define puntos para graficar curva de regresión
x = np.linspace(-3,3, 500);

# Ajusta estilo y tamaño de la figura
plt.style.use('seaborn-poster')
plt.figure(figsize = (12, 8))

# Grafica datos y curva de regresión
plt.plot(xdata, ydata, 'bo', color='black', label='Datos')
plt.plot(x, func(x, *popt), 'r-', label='y = ax^3 + b e^(cx) : \
a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
