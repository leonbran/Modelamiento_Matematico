# Calcula regresion lineal

```python
# Importa librerías necesarias
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

# Define los datos
x = np.array([-2.18,-1.527,-0.61,0.84,1.75,2.38])
y = np.array([14.765,-3.142,-9.252,-10.342,-2.371,19.705])

# Calcula los parámetros de la recta de regresión
res = stats.linregress(x,y)
a = res.slope # Pendiente de la recta
b = res.intercept # Intercepto de la recta con el eje vertical

# Ajusta tamaño de la figura
plt.style.use('seaborn-poster')
plt.figure(figsize = (12, 8))

# Grafica datos y recta de mínimos cuadrados
x_line = np.arange(min(x), max(x), 1)
plt.plot(x, y, 'bo', color='black', label='Datos')
plt.text(0.2, 4.5, 'y = %.5f * x + %.5f' % (a, b), fontsize=15)
plt.plot( x_line, a*x_line + b, color = 'red' )
plt.legend()
plt.show()
```
