# Gr√°fica logaritmo de y versus x

```python
# Importa librerias necesarias
import matplotlib.pyplot as plt
import numpy as np

# Define la funcion para graficar
x = np.linspace(0, 4, 1000)
k = 3.5
y = k*x
a = [-1,4]
b = [0,0]

# Ajusta estilo y tama~no figura
plt.style.use('seaborn-poster')
plt.figure(figsize = (12, 8))

# Crea el grafico
plt.plot(x, y, color='black', linewidth=2)
plt.axvline(x=0, color='black', linewidth=1)
plt.plot(a,b,'-', color='black', linewidth=1)
plt.text(2.5, 6, r'$z = kx$', fontsize=15)
plt.text(2.5, -1.5, r'$x$', fontsize=15)
plt.text(-0.25, 10, r'$z = ln(y)$', fontsize=15)
plt.axis('off')
plt.xlim([-1, 6])
plt.ylim([-1, 20])
plt.xticks([])
plt.yticks([])
plt.show()
```
