# Gr√°fica logaritmo de y versus logaritmo de x

```python
# Importa librerias necesarias
import matplotlib.pyplot as plt
import numpy as np

# Define la funcion a graficar
x = np.linspace(0, 4, 1000)
k = 1.5
y = k*x
a = [-1,4]
b = [0,0]

# Ajusta estilo y tama~no figura
plt.style.use('seaborn-poster')
plt.figure(figsize = (12, 8))

# Crea el grafico
plt.plot(x, y, color='black', linewidth=2)
plt.axvline(x=0, color='black', linewidth=1)
plt.plot(a,b,'-', color='black', linewidth=1)
plt.text(2.5, 6, r'$z = \alpha t$', fontsize=15)
plt.text(1.75, -1.5, r'$t = ln(x)$', fontsize=15)
plt.text(-0.25, 10, r'$z = ln(y)$', fontsize=15)
plt.axis('off')
plt.xlim([-0.25, 5])
plt.ylim([-1, 9.5])
plt.xticks([])
plt.yticks([])
plt.show()
```
