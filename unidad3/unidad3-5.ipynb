# Dibuja gr√°fica exponencial

```python
# Importa librerias necesarias
import matplotlib.pyplot as plt
import numpy as np
# Define la funcion para graficar
x = np.linspace(-2, 5, 1000)
k = 3
y = np.exp(0.8*x)
a = [-1,6]
b = [0,0]
# Ajusta estilo y tama~no figura
plt.style.use('seaborn-poster')
plt.figure(figsize = (12, 8))

# Crea el grafico
plt.plot(x, y, color='black', linewidth=2) # Grafica los datos x, y
plt.axvline(x=0, color='black', linewidth=1) # Traza el eje vertical
plt.plot(a,b,'-', color='black', linewidth=1) # Traza el eje horizontal
plt.text(3.5, 10, r'$y = e^{kx}$', fontsize=15) # Pone la etiqueta y = e^(kx)
plt.axis('off') # Elimina la caja que delimita el grafico por defecto
plt.text(2.5, -1.5, r'$x$', fontsize=15) # Coloca nombre al eje horizontal
plt.text(-0.25, 10, r'$y$', fontsize=15) # Coloca nombre al eje vertical
plt.xlim([-1, 6]) # Define limites del eje horizontal
plt.ylim([-1, 20]) # Define limites del eje vertical
plt.xticks([]) # Para que no aparezca escala en el eje horizontal
plt.yticks([]) # Para que no aparezca escala en el eje vertical
plt.show() # Muestra el grafico en una ventana
```
