# Gráfica para visualizar relación de proporcionalidad

```python
# Importa librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

#Genera la recta a graficar
x = np.linspace(0, 5, 100)
k = 4
y = k*x

#Genera puntos para graficar líneas punteadas
a = [1,1]
b = [0,4]
c = [0,1]
d = [4,4]

# Ajusta tamaño figura
plt.style.use('seaborn-poster')
fig = plt.figure(figsize = (10, 5))

#Crea el gráfico
plt.axvline(x=0, color='black', linewidth=1) # Traza el eje vertical
plt.plot(a,b,'-', color='black', linewidth=1) # Traza el eje horizontal
plt.plot(x, y, color='black') # Grafica los datos x, y
plt.plot(a,b,'--', color='black') # Grafica linea punteada horizontal
plt.plot(c,d,'--', color='black') # Grafica linea punteada vertical

plt.text(1.8, 5, r'$y = kx$, con $k>0$', fontsize=15) #Escribe una etiqueta
#plt.title('Proporcionalidad directa') # Posible titulo de la grafica

plt.xlabel('x') #Coloca nombre del eje horizontal
plt.ylabel('y') #Coloca nombre del eje vertical

plt.xticks([1]) #Coloca una marca en el eje horizontal
plt.yticks([0,4],["0","k"]) #Coloca marcas en el eje vertical

plt.xlim([0, 3]) # Define limites del eje horizontal
plt.ylim([0, 10]) # Define limites del eje vertical

plt.xticks([]) # Para que no aparezca escala en el eje horizontal
plt.yticks([]) # Para que no aparezca escala en el eje vertical
plt.show() # Muestra el gráfico en una ventana
```

```python
# Importa librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Ajusta estilo y tamaño figura
plt.style.use('seaborn-poster')
plt.figure(figsize = (12, 8))

#Genera puntos a graficar
x = np.array([1.5,3.7,4.2,6.1,7.8])
y = np.array([0.5,2.9,3.4,7.5,12.0])


#Crea gráfica
plt.axvline(x=0, color='black', linewidth=1) # Traza el eje vertical
#plt.plot(a,b,'-', color='black', linewidth=1) # Traza el eje horizontal
plt.plot(x, y, 'bo', color='black') #Grafica los puntos
plt.xlabel('x') # Coloca nombre al eje horizontal
plt.ylabel('y') # Coloca nombre al eje vertical
plt.xlim([0, 8]) # Define limites del eje horizontal
plt.ylim([0, 13]) # Define limites del eje vertical
plt.show() # Muestra el gráfico en una ventana
```
