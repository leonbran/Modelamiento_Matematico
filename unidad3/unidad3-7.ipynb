# Gráfica función exponencial

```python
# Define librerias necesarias
import matplotlib.pyplot as plt
import numpy as np

# Define la funcion para graficar
x = np.linspace(0, 5, 1000)
k = 3
y = x**k
a = [-1,6]
b = [0,0]

# Ajusta estilo y tama~no figura
plt.style.use('seaborn-poster')
plt.figure(figsize = (12, 8))

# Crea el grafico
plt.plot(x, y, color='black', linewidth=2)
plt.axvline(x=0, color='black', linewidth=1)
plt.plot(a,b,'-', color='black', linewidth=1)
plt.text(2, 6, r'$y = x^{\alpha}$', fontsize=15)
plt.text(1.5, -1.2, r'$x$', fontsize=15)
plt.text(-0.15, 5, r'$y$', fontsize=15)
plt.axis('off')
plt.xlim([-0.5, 3])
plt.ylim([-1, 10])
plt.xticks([])
plt.yticks([])
plt.show()
```
