# Interpolación de Lagrange

#### Polinomios de interpolación de Lagrange 

```python
#importamos el modulos de numpy (numeric python)
import numpy as np

#importamos del módulo matplotlib la libreria pyplot con el alias de plt 
import matplotlib.pyplot as plt


#-----------------------------------------------------
#Definiendo los polinomios de lagrange Li(x)"""
def L( x , x_datos ,  i):
     
    #inicializamos a 1    
    Li = 1

    #hago la productoria
    for k in range( len( x_datos ) ): 
                
        #adiciono la restriccion de la productoria
        if k != i:
            Li  = Li * (x- x_datos[k]) / ( x_datos[i]- x_datos[k] )      

    return Li  
#-----------------------------------------------------


#graficando algunos polinomio de lagrange

#numero de polinomios a construir
#datos 
x_datos = [ -1 , 0 , 1 ]


#dominio en donde los graficaremos
x = np.linspace( -1, 1 , 20)



#Graficando los polinomios juntos
for i in range( len( x_datos ) ):

    plt.plot(  x , L( x , x_datos ,  i) , label = "L"+str(i)+"(x)" )

    #poniendo nombres a de los ejes
    plt.xlabel('x')
    plt.ylabel('Li(x)')
    
#mostrando las legendas del grafico
plt.legend()
plt.grid()       
    
#Mostrando el gráfico
plt.show()

      
    
```

#### Construyendo un polinomio de interpolación

```python
#--------------------------------------------
#DATOS
x_datos  = [-1 ,1, 2]
f_datos  = [ 1 ,4, 2] 
#--------------------------------------------

    
#--------------------------------------    
#Definiendo el polinomio de interpolacion de Lagrange
def P(x): 
    
    Suma = 0 
    for i in range( len(x_datos) ):

        Suma = Suma + f_datos[i] * L( x , x_datos ,  i)
   
    return Suma
#--------------------------------------


#hallamos algunas aproximaciones
print("\n")
print( "P(-1) = " , P(-1) , ", lo que coincide con ", f_datos[0] , "\n")
print( "P( 1) = " , P( 1) , ", lo que coincide con ", f_datos[1] , "\n")
print( "P( 2) = " , P( 2) , ", lo que coincide con ", f_datos[2] , "\n")


#aproximamos otros valores
print( "P( 0) = " , P( 0) , ", lo que coincide con ", 11. / 3. , "\n")
print( "P( 3) = " , P( 3) , "\n" )
print( "P( 4) = " , P( 4) , "\n" )
    
```

#### Graficando el polinomio de interpolación y los datos

```python

#dominio en donde los graficaremos
x = np.linspace( -1.5, 2 , 40)


#Graficando las parejas de datos
plt.plot(  x_datos , f_datos , "or" , label="datos" )



#Graficando el polinomio de interpolacion
plt.plot(  x , P(x) , label="polinomio"  )


#poniendo nombres a de los ejes
plt.xlabel('x')
plt.ylabel('P(x)')

#mostrando las legendas del grafico
plt.grid()       
plt.legend()
    
#Mostrando el gráfico
plt.show()
```
