###  Proporcionalidad inversa:

```python
import numpy as np
import matplotlib.pyplot as plt


#los datos 
x = np.linspace(1, 10    , 20)
y = 3/x
        
    
#Graficando las dos listas de valores
plt.plot(x,y)

#poniendo nombres a de los ejes
plt.xlabel('x')
plt.ylabel('y')

#nombre del gráfico
plt.title('grafico x contra y')
 
#Mostrando el gráfico
plt.show()

```

##### Convirtiendo a una relacion lineal

Si $y=c/x$ entonces hacemos el cambio de variable $z=1/x$ para obtener la relacion lineal

$$ y = c z $$

```python
#Cambiando de variable
z = 1/x

#Graficando las dos listas de valores
plt.plot(z,y)

#poniendo nombres a de los ejes
plt.xlabel('z')
plt.ylabel('y')

#nombre del gráfico
plt.title('grafico z contra y')
 
#Mostrando el gráfico 
plt.show()
```

##### Estimando el valor de la constante de proporcionalidad

si $y= c z$  entonces
$$ C = y/z$$

```python
#tomando el promedio
C = np.mean( y / z )

print("el valor de la constante de proporcionalidad inversa es: ",C)
    
```





### Proporcionalidad exponencial

Aqui la relacion de los datos es : $ y = e^{C x} $

```python
import numpy as np
import matplotlib.pyplot as plt


#--------------------------------------------
#DATOS
#--------------------------------------------
x = np.linspace(1, 10    , 20)
y = np.exp( 0.5 * x)
 
         
#Graficando las dos listas de valores
plt.plot(x,y)

#poniendo nombres a de los ejes
plt.xlabel('x')
plt.ylabel('y')

#nombre del gráfico
plt.title('x contra y')
 
#Mostrando el gráfico
plt.show()
```

##### Convirtiendo a una relacion lineal

Debe tomarse logaritmo a ambos lados $$\text{Log}y = C x $$

Hacemos el cambio de variable $z=\text{Log}y$ para obtener la recta 

$$z = C x$$

```python
#Cambio de variable
z = np.log(y)


#Graficando las dos listas de valores
plt.plot(x,z)


#poniendo nombres a de los ejes
plt.xlabel('x')
plt.ylabel('log(y)')
#nombre del gráfico
plt.title('log(y) contra x')
 
#Mostrando el gráfico
plt.show()
```

##### Estimando el valor de la constante de crecimiento

Si   $z = C x $ entonces

$$C =z / x$$. 

```python
#encontrando la constante de proporcionalidad
C = z/x

#tomando el promedio
Constante = np.mean(C)

print("Valor aproximado de la constante de crecimiento = ", Constante)
```






### Proporcionalidad de potencia o petencial 

Aqui los datos se relacionan como $y = C x^{ n }$

```python
import numpy as np
import matplotlib.pyplot as plt

 
#--------------------------------------------
#DATOS
#--------------------------------------------
x = np.linspace(2, 20    , 20)
y = 0.1* x**(-2) 
 
        
#Graficando las dos listas de valores
plt.plot(x,y)

#poniendo nombres a de los ejes
plt.xlabel('x')
plt.ylabel('y')
#nombre del gráfico
plt.title('x contra y')
 
#Mostrando el gráfico
plt.show()
```

###### Conviertiendo a una relacion lineal

Si $y = c x^{ n }$ entonces: 

\begin{eqnarray}
\text{Log} y &=& \text{Log} c x^{ n }\\ 
 \text{Log} y            &=&  \text{Log} x^{ n } +  \text{Log} c \\
  \text{Log} y           &=& n \text{Log} x +  \text{Log} c
\end{eqnarray}

Entonces definimos las variables  $z = \text{Log} x$, $w = \text{Log} y$ y $b = \text{Log}c $ para obtener la recta

$$ w = n z + b $$



```python
#Cambio de variable
z = np.log(x)
w = np.log(y)

#Graficando las dos listas de valores
plt.plot(z,w)

#poniendo nombres a de los ejes
plt.xlabel('log(x)')
plt.ylabel('log(y)')
#nombre del gráfico
plt.title('log(x) contra log(y)')
 
#Mostrando el gráfico
plt.show()
```

#### Estimando el valor de las constantes ?

Si $ w = n z + b $ entonces:

- $ n $ es la pendiente de la recta
- $ b $ el intercepto de la recta.   

Tenemos **una** ecuacion y **dos** variables. Como podemos estimar el valor de las variables?

### 3. Interpolación de Lagrange (en caso de tener tiempo)

Usando el modulo scipy podemos el poliomio $P(x)$ que interpola los n-datos $(x_i,y_i)$ usando el metodo de Lagrange.  Recordar que el polinio interpolante de Lagrange se define como

$$P(x) = \sum^{n}_{i=0} y_i L_i(x)$$

En donde los $L_i(x)$ son los polinomios de Lagrange definidos como:

$$L_i(x) = \prod^{n}_{j=0} \dfrac{x-x_j}{x_i-x_j}$$


**Ejemplo**

```python
import numpy as np
import matplotlib.pyplot as plt

#Nuevo comando!
from scipy.interpolate import lagrange


#puntos en x
x = np.array([-1,    0,  1])

#puntos en y
y = np.array([5.4, 4.2, -3])


#grafico de los puntos
plt.plot( x , y , 'ro' , label='puntos')


#polinomio de lagrange( es una funcion de numpy )!!!
PolinomioDeLagrange = lagrange(x, y)

#dominio entre -1 y 1 para evaluar el polinomio
dominiox = np.linspace(-1, 1, 20)

plt.plot( dominiox , PolinomioDeLagrange(dominiox) , 'b' , label='polinomio interpolante')


#adicionando cosas al grafico
plt.legend()
plt.grid()


#mostrando el grafico
plt.show()
```
