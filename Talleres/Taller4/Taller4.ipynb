**Ejemplo:** En este programa se muestra como graficar la funci√≥n ùë¶(ùë•)=  exp( cos(x) ) en el dominio [0,2ùúã] mediante la implementaci√≥n del m√≥dulo numpy y el m√≥dulo matplolib. La idea clave es utilizar arreglos tipo numpy para definir el dominio de una funciones y evaluarla sobre este arreglo de numeros. 

```python
#importamos el modulos de numpy (numeric python)
import numpy as np


#importamos del m√≥dulo matplotlib la libreria pyplot con el alias de plt 
#(esto no permitira graficar las funciones).
import matplotlib.pyplot as plt


#numero de divisiones del intervalo
n = 100

#--------------------------------------------
#defino el dominio como un arreglo tipo numpy
#--------------------------------------------
x = np.linspace(0, 2*np.pi, n)


#mostrar la lista de los valores de x
print("Los valores de la variable x son:\n ", x, "\n")



#Mediante el m√≥dulo numpy, evaluo la funci√≥n sin(x)+ cos(x) en todo el dominio
#pasando el arreglo tipo numpy de x a la funcion. Esto arrojara 
#una lista de valores como resultado
#--------------------------------------------


def MiFuncionCompuesta(x):
    
    f = np.cos(x)
    g = np.exp( f )   
    
    return g



y = MiFuncionCompuesta(x)


#mostrar la lista de los valores de y
print("Los valores de la variable y son:\n ",   y, "\n")


#Graficando las dos listas de valores
plt.plot(x,y)


#poniendo nombres a de los ejes
plt.xlabel('x')
plt.ylabel('y')


#nombre del gr√°fico
plt.title('Grafico de y=sin(x)')

# ajustar tama√±o de la figura (ancho y alto)
plt.rcParams['figure.figsize'] = [8, 6]
 
#Mostrando el gr√°fico
plt.show()
```

### Guardando los datos de la funcion en un archivo

```python
# hacer dos columnas, crear un arreglo
datosDosColumnas = np.column_stack(  (x, y)  )
print(datosDosColumnas)
 

# guardado de numpy!!!!
np.savetxt('archivo.dat', datosDosColumnas, header = "X-Columna, Y-Columna")
print ("listo!")
```
